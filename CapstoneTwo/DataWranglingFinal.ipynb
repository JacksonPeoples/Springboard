```python
# One important difference is the time format. Will either need to create a column for local time on the 
# weather data or a utc column on the crash data.

#  Definitely some columns in crash data that won't be helpful. 'case_id' is unuseful because we already have 
# 'crash_id' to differentiate between evetns. The rest because of either redundancy or lack of data.

del_col = ['case_id', 'rpt_latitude', 'rpt_longitude', 'units_involved','atd_mode_category_metadata', 'street_nbr_2']
crash_data = crash_data.drop(del_col, 1)
```

```python
# to match weather data will need to create a column in crash data that is UTC unix. 
# First need to convert 'crash_time' to datetime.

crash_data['crash_date']= pd.to_datetime(crash_data['crash_date'])
```

```python
#need to also round to hour to match granularity of weather data.

def rounding(time):
    return time.floor('H')

crash_data['utc_unix'] = crash_data['crash_date'].apply(rounding)
```

```python
# Probably not the most elegant way, but will write out what times are daylight savings, then convert to unix.

def utc(time):
    if time > pd.Timestamp('2010-03-14T02') and time < pd.Timestamp('2010-11-07T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2011-03-13T02') and time < pd.Timestamp('2011-11-06T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2012-03-11T02') and time < pd.Timestamp('2012-11-04T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2013-03-10T02') and time < pd.Timestamp('2013-11-03T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2014-03-09T02') and time < pd.Timestamp('2014-11-02T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2015-03-08T02') and time < pd.Timestamp('2015-11-01T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2016-03-13T02') and time < pd.Timestamp('2016-11-06T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2017-03-12T02') and time < pd.Timestamp('2017-11-05T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2018-03-11T02') and time < pd.Timestamp('2018-11-04T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2019-03-10T02') and time < pd.Timestamp('2019-11-03T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2020-03-08T02') and time < pd.Timestamp('2020-11-01T02'):
        new_time = time + pd.Timedelta('5 hours')
    else:
        new_time = time + pd.Timedelta('6 hours')
    return new_time
        
def unix(time):
    return (time - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')
    
def unix_utc(time):
    return unix(utc(time))
```

```python
crash_data['utc_unix'] = crash_data['utc_unix'].apply(unix_utc)
```

```python
# now I need to checkout the weather data, the biggest issue in merging would be if there were more than one
# weather observation per hour.

print(len(weather_data['dt_iso']))
print(weather_data['dt_iso'].nunique())
```

```python
# there does appear to be more than one observation per hour. I'll group by each column and see which columns result
# in more groupings than we have dates. That will tell me what the repeat times are differentiating on.

for col in weather_data.columns:
    if weather_data['dt_iso'].nunique() < weather_data.groupby(['dt_iso',col]).ngroups:
        print(col)
        print(weather_data.groupby(['dt_iso',col]).ngroups)
```

```python
# based on documentation, these aren't very important columns. They're created as a function of the numbers
# recorded in other columns to offer a display or summary. If I drop these and then delete repeat rows, I should
# have only unique observations.

weather_cols_drop = ['weather_id', 'weather_main', 'weather_description', 'weather_icon']
weather_data = weather_data.drop(weather_cols_drop, 1)
```

```python
weather_data = weather_data.drop_duplicates()
```

```python
# Success! Now should be theoretically ready to merge these two dataframes.
weather_data = weather_data.rename(columns= {'dt':'utc_unix'})
```

```python
df = pd.merge(crash_data, weather_data, how= 'left', on = 'utc_unix')
```
