```python
#need to also round to hour to match granularity of weather data.

def rounding(time):
    return time.floor('H')

crash_data['utc_unix'] = crash_data['crash_date'].apply(rounding)
```

```python
# Probably not the most elegant way, but will write out what times are daylight savings, then convert to unix.

def utc(time):
    if time > pd.Timestamp('2010-03-14T02') and time < pd.Timestamp('2010-11-07T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2011-03-13T02') and time < pd.Timestamp('2011-11-06T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2012-03-11T02') and time < pd.Timestamp('2012-11-04T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2013-03-10T02') and time < pd.Timestamp('2013-11-03T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2014-03-09T02') and time < pd.Timestamp('2014-11-02T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2015-03-08T02') and time < pd.Timestamp('2015-11-01T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2016-03-13T02') and time < pd.Timestamp('2016-11-06T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2017-03-12T02') and time < pd.Timestamp('2017-11-05T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2018-03-11T02') and time < pd.Timestamp('2018-11-04T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2019-03-10T02') and time < pd.Timestamp('2019-11-03T02'):
        new_time = time + pd.Timedelta('5 hours')
    elif time > pd.Timestamp('2020-03-08T02') and time < pd.Timestamp('2020-11-01T02'):
        new_time = time + pd.Timedelta('5 hours')
    else:
        new_time = time + pd.Timedelta('6 hours')
    return new_time
        
def unix(time):
    return (time - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')
    
def unix_utc(time):
    return unix(utc(time))
```

```python
crash_data['utc_unix'] = crash_data['utc_unix'].apply(unix_utc)
```
